## Important packages

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import L2

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, mean_absolute_error
```

## Data normaliation

```python
scaler = StandardScaler()
training_data = scaler.fit_transform(data[relevant_columns])
```

## Data Formating to tensor

```python
import numpy as np

n_past = 15
n_future = 1

m,n = training_data.shape

final_trainX_dim = (m - n_past - n_future+1, n_past, n)
#final_trainY_dim = (m - n_past - n_future, n_future, n)

trainX = np.zeros(final_trainX_dim)
trainY = training_data[n_past:,:]

for i in range(m-n_past-n_future):
  trainX[i] = training_data[i:i+n_past, :]

```

```python
print(trainX.shape)
print(trainY.shape)
```

```python
training_data[15]
```

```python
trainX[15]
```

```python
n_feature = n
```

```python
model =Sequential([
    LSTM(units=50, input_shape=(n_past,n_feature), return_sequences=True, kernel_regularizer = L2(0.01)),
    Dropout(0.4),
    LSTM(units =25, return_sequences = True,kernel_regularizer = L2(0.01)),
    Dropout(0.4),
    LSTM(units = 10, return_sequences = False, kernel_regularizer = L2(0.01)),
    Dropout(0.4),
    Dense(units = n_feature)
])

#Model compilation by speifying loss and optimizer
model.compile(optimizer = Adam(0.01), loss = 'mse')
model.summary()
```

```python
history = model.fit(trainX.astype('float32'), trainY.astype('float32'), epochs=32, validation_split=0.2, batch_size = 32)
```

```python
history.history
```

```python
fig3, ax3 = plt.subplots(1,1, figsize = (8,5))
fig3.tight_layout(pad = 2)
ax3.plot(history.history['loss'])
ax3.plot(history.history['val_loss'])
ax3.legend(['loss', 'val loss'])
```
